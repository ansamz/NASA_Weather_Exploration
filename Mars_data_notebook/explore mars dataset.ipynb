This dataset contains all of the data used in The Pudding essay Greetings from Mars. It captures the weather conditions on Mars at Gale Crater from Sol 1 to Sol 1895. This data was measured by and transmitted from the Rover Environmental Monitoring Station (REMS) on-board the Curiosity Rover. All environmental magnitudes provided are obtained from values reported back to Earth and may be impacted by Curiosity's position, temperature, orientation, shade or dust depositions.

This folder also includes weather information compiled at UTC Midnight for 1,000 cities around the world - ranging from capital cities of every country to states within the U.S. This data is Powered by Dark Sky with each row containing latitude/longitude coordinates alongside a 3-day window of forecasted/predicted regions for each respective city/state.

Both files come with a disclaimer as outlined above and must include attribution as Powered by Dark Sky should you use this dataset for further analysis purposes - released under an MIT License agreement!

The information in this file is outlined in different columns - there are 14 columns listed: id, terrestrial_date, sol, ls, month,min_temp, max_temp,pressure ,wind_speed ,atmo_opacity. As well as specific descriptions provided that contain more detail as what each column contains(For example: variable called “min_temp” includes temperatures that were observed lower than average during a single Martian day). Additionally, there are coordinates included within each row - latitude and longitude which provides physical locations where these measurements were taken at.

About mars-weather.csv

The mars-weather.csv file contains data on Martian sols, minimum and maximum temperatures, atmospheric pressure, wind speeds, and atmospheric opacity from 2012 to 2018

- terrestrial_date: The date on Earth when the data was collected. (Date)
- sol: The Martian day when the data was collected. (Integer)
- ls: The Martian season when the data was collected. (String)
- min_temp: The minimum temperature on Mars when the data was collected. (Float)
- max_temp: The maximum temperature on Mars when the data was collected. (Float)
- pressure: The atmospheric pressure on Mars when the data was collected. (Float)
- wind_speed: The wind speed on Mars when the data was collected. (Float)
- atmo_opacity: The atmospheric opacity on Mars when the data was collected. (String)
- month: The month when the data was collected. (String)
