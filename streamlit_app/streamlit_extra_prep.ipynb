```python
import pandas as pd
import db_connect
import os
```

```python
earth_weather_data = db_connect.earth_weather_data()
```

```python
if not os.path.exists("data"):
    os.makedirs("data")
```

```python
earth_weather_data["date"] = pd.to_datetime(earth_weather_data["date"])

earth_weather_data["date"] = earth_weather_data["date"].dt.normalize()

# saving each locations data to a parquet file (1 observation per day, either first or mean)
agg_methods = {
    "latitude": "first",  
    "longitude": "first",  
    "temperature_2m": "mean", 
    "relative_humidity_2m": "mean",  
    "rain": "mean", 
    "direct_radiation_instant": "mean" 
}

for location in earth_weather_data["location"].unique():
    subset = earth_weather_data[earth_weather_data["location"] == location]
    grouped_subset = subset.groupby("date").agg(agg_methods).reset_index()
    grouped_subset.to_parquet(os.path.join("data", f'{location}.parquet.gzip'), index=False, compression="gzip")
    grouped_subset.to_csv(os.path.join('data', f'{location}.csv'), index=False)
```
